```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
df = pd.read_csv('12042017074931-data.csv')
df.info()
```

```python
key = {'nf':'number of fish','ns':'number of sharks','fs':'mean fish speed','fa':'mean fish angular speed',
      'fcw':'fish cohesion weight','faw':'fish alignment weight','fsew':'fish separation weight',
      'fshw':'fish shark weight','fdw':'fish dodge weight','ffw':'fish food weight','ss':'mean shark speed',
      'sa':'mean shark angular speed','ssw':'shark separation weight','sfw':'shark fish weight','ssize':'shark size'}
```

```python
file = open('12042017074931-info.csv','r')
keys = file.readline().split(',')
values = file.readline().split(',')
file.close()
deviations = dict()
for value in key:
    if not value in ['nf','ns']:
        deviations[value] = values[keys.index('sd'+value)]
print(deviations)
```

```python
fig = plt.figure(figsize=(10,20))
i = 1
for param in df.columns[1:]:
    ax = fig.add_subplot(8,2,i)
    ax.plot(df['time'],df[param])
    if not param in ['nf','ns']:
        ax.set_title(key[param] + ' vs time $(\sigma=' + deviations[param] + ')$')
    else:
        ax.set_title(key[param] + ' vs time')
    i += 1
plt.tight_layout()
plt.savefig('12042017074931.png')
plt.show()
```
