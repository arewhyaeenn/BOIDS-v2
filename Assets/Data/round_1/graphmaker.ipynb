```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('12022017173427-data.csv',
                #names=['time', 'nf', 'ns', 'fs', 'ss', 'fa', 'sa'],
                dtype={'time':float, ' nf':int, ' ns':int, ' fs':float,
                      ' ss':float, ' fa':float, ' sa':float})
df.info()
```

```python
fig = plt.figure(figsize=(10,5))
ax1 = fig.add_subplot(121)
ax2 = fig.add_subplot(122)
T = df['time']
NF = df[' nf']
NS = df[' ns']
FS = df[' fs']
SS = df[' ss']
#FA = df[' fa']
#SA = df[' sa']
ax2.plot(T,NF,label='number of fish')
ax1.plot(T,NS,label='number of sharks')
ax1.plot(T,FS*10,label='mean fish speed')
ax2.plot(T,SS*10,label='mean shark speed')
ax2.set_ylim(0.20)
ax1.legend()
ax2.legend()
plt.show()
```
